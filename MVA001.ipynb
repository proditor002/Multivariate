#Eigen value กับ Eigen vector¶
Eigen value ($\lambda$) -> $|A-\lambda I| = 0$

Eigen vector ($x$) -> $Ax = \lambda x$

#eigen vector and eigen of covarience matrix of data

eigen vevtors and eigen values
จากรูป มีจุดข้อมูล 16 จุด และมี aatribute 2มิติ เราจะสร้างเป็น data matrix ที่มีรูปร่าง (shape)? = (2,16)

```
 X = np.array([[1,2,-3,4],[4,5,-3,2]])
 print(f'data X =\n {X}')
 C_X = np.cov(X)
 print(f'Covarience ของ X คือ \n {C_X}')
```

```
W,V = np.linalg.eig(C_X)
```

```
W
```

```
print(W[1])
print(V[:,1])
```

```
plt.plot(X[0,:],X[1,:],'or')
plt.plot([0,3*V[0,1]],[0,3*V[1,1]],'<-b')
```

#INVERSE MATRIX 
inverse ของ $A$ เขียน $A^{-1}$
นิยาม $A*A{-1} = I = A^{-1}*A$

ลิ้ง https://numpy.org/doc/stable/reference/generated/numpy.linalg.inv.html

```
C_X
```

```
inv_C_X = np.linalg.inv(C_X)
print(inv_C_X)
```

ตรวจสอบว่าตรงกับนิยามมั้้ย

```
np.dot(C_X,inv_C_X)
```

```
np.dot(inv_C_X,C_X)
```

```
np.linalg.inv(inv_C_X)
```
